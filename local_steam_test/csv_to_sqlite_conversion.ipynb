```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
engine=create_engine('sqlite:///assests/data/games.sqlite')
```

```python
pd.read_csv('assests/data/steam_sales_by_country.csv').drop('Unnamed: 0', axis=1).to_sql('country_sales', engine.connect())
```

```python
conn=engine.connect()
```

```python
results=conn.execute('select * from country_sales;').fetchall()
list(results)
```

```python
results_lol=[list(result) for result in results]
results_lol
```
