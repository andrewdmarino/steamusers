```python
url_first = "HTML Files/html-state-"

url_second = ".html"

counter = 0

list_of_states = []

list_of_values = []

for states in state_abbreviations:
    state_value = 0
    soup = BeautifulSoup(open(url_first + states + url_second, encoding='utf8'), "html.parser")
    pointer = soup.findAll("td", {"class" : "sorting_1"})
    for value in pointer:
#         print(value.get_text())
        if (value.get_text() == "Private"):
            print("oops")
        else:
            try:
                new_value = value.get_text()
                number_of_games_int = int(new_value)
                state_value = state_value + number_of_games_int
            except:
                new_value = value.get_text()
                newstring = new_value.replace(",","")
                number_of_games_int = int(newstring)
                state_value = state_value + number_of_games_int
    list_of_states.append(state_names[counter])
    list_of_values.append(state_value)
    counter = counter + 1
```

```python
games_by_state = pd.DataFrame()

games_by_state["State"] = list_of_states
games_by_state["Abbreviation"] = state_abbreviations
games_by_state["Games Owned"] = list_of_values
```

```python
games_by_state = games_by_state.sort_values(by="Games Owned", ascending=False)

games_by_state.to_csv("state_data.csv")

games_by_state
```
